```python
import difflib # Import the difflib module
import pickle

# Load the data and similarity matrix
movies_data = pickle.load(open('movies_list.pkl','rb'))
similarity = pickle.load(open('similarity.pkl','rb'))


movie_name= input('Entre your favorite movie name: ')
list_of_all_titles=movies_data['title'].tolist()
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

if find_close_match: # Check if a close match was found
  close_match = find_close_match[0]
  index_of_the_movie = movies_data[movies_data.title == close_match].index[0] # Access index correctly
  sorted_similar_movies = sorted(list(enumerate(similarity[index_of_the_movie])),key = lambda x:x[1],reverse=True) # Use similarity

  print('movies suggested for you :\n')
  i=1
  for movie in sorted_similar_movies:
    index = movie[0] # Use 'movie' instead of 'movies'
    title_from_index= movies_data.iloc[index]['title'] # Access title correctly using iloc
    if (i<30):
      print(i,'.',title_from_index)
      i+=1 # Increment 'i' with i += 1
else:
  print(f"No close match found for '{movie_name}' in the dataset.")
```
