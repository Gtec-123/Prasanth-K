```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/zomato.csv.zip')
data
```

```python
data=data.drop(['url','address','phone','menu_items','dish_liked','reviews_list'],axis=1, errors='ignore')
data.head()
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
data.shape
```

```python
data['rate'].unique()
```

```python
def handlerate(value):
  if(value=='NEW' or value=='_' or value=='-'): # Added condition for '-'
    return np.nan
  else:
    value=str(value).split('/')
    value = value[0]
    return float(value)

data['rate']=data['rate'].apply(handlerate)
data['rate'].head()
```

```python
data['rate'].fillna(data['rate'].mean(), inplace=True)
data['rate'].isnull().sum()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
data.head()
```

```python
data.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type'},inplace=True)
data.head()
```

```python
data['location'].unique()
```

```python
data['listed_in(city)'].unique()
```

```python
data=data.drop(['listed_in(city)'],axis=1)
```

```python
data['cost2plates'].unique()
```

```python
def handlecomma(value):
  value=str(value)
  if ',' in value:
    value=value.replace(',','')
    return float(value)
  else:
    return float(value)
data['cost2plates']=data['cost2plates'].apply(handlecomma)
data['cost2plates'].unique()
```

```python
data.head()
```

```python
rest_types= data['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
  if value in rest_types_lessthan1000:
    return 'others'
  else:
    return value

data['rest_typr'] = data['rest_type'].apply(handle_rest_type)
data['rest_type'].value_counts()
```

```python
# Define the handle_location function
location =data['location'].value_counts(ascending=False)
location_lessthan300=location[location<300]
def handle_location(value):
  if(value in location_lessthan300):
    return'other'
  else:
    return value

# Apply the function to the 'location' column and check value counts
data['location']=data['location'].apply(handle_location)
data['location'].value_counts()
```

```python
cuisines=data['cuisines'].value_counts(ascending=False)
cuisines_lessthan100=cuisines[cuisines<100]

def handle_cuisines(value):
  if(value in cuisines_lessthan100):
    return 'others'
  else:
    return value

data['cuisiness']=data['cuisines'].apply(handle_cuisines)
data['cuisines'].value_counts()
```
